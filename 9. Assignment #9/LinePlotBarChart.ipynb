```python
js_text_template = Template('''
  var margin = {top:20,right:20,bottom:30,left:50};
  var width = 900 - margin.left - margin.right;
  var height = 700 - margin.top - margin.bottom;

  var xScale = d3.scaleBand().range([0, width]).padding(0.4);
  var yScale = d3.scaleLinear().range([height,0]);

  var svg = d3.select("#$maid").append("svg")
  .attr("width", width + margin.left+margin.left)
  .attr("height", height + margin.top+margin.bottom)
  .append("g")
  .attr("transform","translate("+margin.left+","+margin.top+")");

  var data = $python_data;
  
  data.forEach(function(d) {
      d.Year = +d.Year;
      d.Number = +d.Number;
  });

  console.log("Extent",d3.extent(data,function(d) { return d.Year}));
  console.log(data);

  xScale.domain(data.map(function(d) {return d.Year;}));
  yScale.domain([0,d3.max(data, function(d) {return d.Number;})]);

  svg.append("text")
  .attr("x", 300)
  .attr("y", 5)
  .attr("class", "title")
  .text("Marriages in KZ");
  
  svg.append("g")
  .attr("transform","translate(0,"+height+")")
  .call(d3.axisBottom(xScale))
  .append("text")
  .attr("class","label")
  .attr("x", width + 10)
  .style("text-anchor","end")
  .text("Year");

  svg.append("g")
  .call(d3.axisLeft(yScale));

  svg.selectAll(".bar")
  .data(data, (d)=>data)
  .enter()
  .append("rect")
  .attr("class","bar")
  .attr("x", (d)=>xScale(d.Year))
  .attr("y", (d)=>yScale(d.Number))
  .attr("width", xScale.bandwidth())
  .transition()
  .ease(d3.easeLinear)
  .duration(500)
  .delay((d,i)=>i*50)
  .attr("height", (d)=>height-yScale(d.Number));

  console.log(data);
''');
```

```python
html_template = Template('''
    <script src="https://d3js.org/d3.v4.js" charset="utf-8"></script>
    <style>
         .bar {
           fill: #b9a0f2;
         }
         .label {
           fill: black;
           font: 10px Georgia, serif;
         }
         .axis path,
         .axis line{
             fill: none;
             shape-rendering:crispEdges;
         }
         .title{
           fill: black;
           font: 20px Georgia, serif;
         }
    </style>
    <div id="barchart-id"></div>
    <script>$js_text</script>
''')
js_text = js_text_template.substitute({
    "python_data":json.dumps(df_dict),
    "maid":"barchart-id"
})

HTML(html_template.substitute({'js_text':js_text}))
```

```python
js_text_template = Template('''
  var margin = {top:20,right:20,bottom:30,left:50};
  var width = 900 - margin.left - margin.right;
  var height = 600 - margin.top - margin.bottom;

  var xScale = d3.scaleTime().range([0, width]);
  var yScale = d3.scaleLinear().range([height, 0]);

  var svg = d3.select("#$maid").append("svg")
  .attr("width", width + margin.left+margin.left)
  .attr("height", height + margin.top+margin.bottom)
  .append("g")
  .attr("transform","translate("+margin.left+","+margin.top+")");

  var data = $python_data;
  
  data.forEach(function(d) {
      d.Year = d3.timeParse("%Y")(d.Year);
      d.Number = +d.Number;
  });
  
  console.log("Extent",d3.extent(data,function(d) { return d.Year}));
  console.log(data);

  xScale.domain(d3.extent(data, function(d) { return d.Year; }));
  yScale.domain([0, d3.max(data, function(d) {return d.Number+10; })]);

  svg.append("text")
  .attr("x", 300)
  .attr("y", 5)
  .attr("class", "title")
  .text("Marriages in KZ");

  svg.append("g")
  .attr("transform","translate(0,"+height+")")
  .call(d3.axisBottom(xScale))
  .append("text")
  .attr("class","label")
  .attr("x", width + 10)
  .style("text-anchor","end")
  .text("Year");

  svg.append("g")
  .call(d3.axisLeft(yScale));

  svg.selectAll("path")
  .datum(data, (d)=>data)
  .attr("fill", "none")
  .attr("stroke", "#eb8dae")
  .attr("stroke-width", 3)
  .attr("d", d3.line()
        .x(function(d){
          return xScale(d.Year);
        })
        .y(function(d){
          return yScale(d.Number);
        })
  );

  console.log(data);
  
''');
```
